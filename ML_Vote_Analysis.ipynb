### Cropping Gambar dari Data Test

```python
from ultralytics import YOLO
import cv2
import os

# Memuat model YOLOv8
model = YOLO('/content/runs/detect/train/weights/best.pt')

# Menetapkan direktori untuk input gambar
input_dir = '/content/Test'

# Menetapkan direktori output gambar hasil cropping akan disimpan
output_dir = 'cropped_images'

# Membuat direktori output jika belum ada
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Looping semua gambar di direktori masukan
for filename in os.listdir(input_dir):
    if filename.endswith('.jpg'):
        image_path = os.path.join(input_dir, filename)
        # Melakukan inferensi pada gambar
        results = model(image_path)
        # Memuat gambar asli
        img = cv2.imread(image_path)
        # Mengambil kotak pembatas dan label kelas
        boxes = results[0].boxes.xyxy.tolist()
        labels = results[0].boxes.cls.tolist()
        # Mengulangi kotak pembatas dan label kelas
        for i, (box, label) in enumerate(zip(boxes, labels)):
            x1, y1, x2, y2 = box
            # Memotong objek menggunakan koordinat kotak pembatas
            ultralytics_crop_object = img[int(y1):int(y2), int(x1):int(x2)]
            # Mendapatkan nama objek terdeteksi dari objek hasil
            object_name = results[0].names[label]
            # Mendapatkan nama dasar dari gambar masukan
            basename = os.path.basename(image_path)
            basename, _ = os.path.splitext(basename)
            # Membangun nama file dengan format yang diinginkan
            filename = f'{basename}_{object_name}.jpg'
            # Membangun path file
            filepath = os.path.join(output_dir, filename)
            # Menyimpan objek yang sudah dicrop sebagai gambar dengan nama file baru
            cv2.imwrite(filepath, ultralytics_crop_object)
```

### Prosedur OMR

```python
# Muat gambar dan ubah ukurannya agar memiliki tinggi 500 piksel
folder_path = '/content/cropped_images'
for filename in os.listdir(folder_path):
    if filename.startswith('TPS') and filename.endswith('.jpg') and ('omr1' in filename or 'omr2' in filename or 'omr3' in filename):
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        # Periksa jika gambar ini terbalik, jika terbalik akan dirotate
        if '-kebalik' in filename:
            image = cv2.rotate(image, cv2.ROTATE_180)

        height, width, _ = image.shape
        aspect_ratio = width / height
        new_height = 500
        new_width = int(new_height * aspect_ratio)
        image = cv2.resize(image, (new_width, new_height))

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

        kernel = np.ones((5, 5), np.uint8)
        closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
        contours, _ = cv2.findContours(closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        image_with_contours = cv2.cvtColor(closing, cv2.COLOR_GRAY2BGR)

        num_columns = 3 # Banyak kolom pada gambar untuk di OMR (yakni 3)
        num_rows = 10 # Banyak baris pada gambar (digit 0 sampai 9)
        answers = [[] for _ in range(num_columns)]
        expected_area = 1000

        print(f"Filename: {filename}")

        for contour in contours:
            area = cv2.contourArea(contour)
            if area > expected_area:
                x, y, w, h = cv2.boundingRect(contour)
                column_index = int(round(x / (gray.shape[1] / num_columns)))
                row_index = int(round(y / (gray.shape[0] / num_rows)))

                # Periksa jika index kolom berada dalam jangkauan yang valid
                if 0 <= column_index < len(answers):
                    if len(answers[column_index]) == 0:
                        answers[column_index].append(row_index)
                    else:
                        last_idx = answers[column_index][-1]
                        try:
                            if area > cv2.contourArea(contours[last_idx]):
                                answers[column_index] = [row_index]
                            elif row_index != last_idx:
                                answers[column_index].append(row_index)
                        except IndexError:
                            # Tangani kasus di mana last_idx berada di luar jangkauan untuk kontur
                            print(f"Error processing contour for {filename}: index {last_idx} out of range.")
                else:
                    print(f"Column index {column_index} is out of valid range for {filename}")
                cv2.drawContours(image_with_contours, [contour], -1, (0, 255, 0), 2)

        for index, answer_list in enumerate(answers):
            print(f"Column {index + 1}: Number marked - {answer_list}")
```

```python
# Prosedur OMR sehingga menghasilkan output berupa dataframe
import cv2
import numpy as np
import os
import pandas as pd

# Muat gambar dan ubah ukurannya agar memiliki tinggi 500 piksel
folder_path = '/content/cropped_images'
data = {'TPS': [], 'Column 1': [], 'Column 2': [], 'Column 3': []}

for filename in os.listdir(folder_path):
    if filename.startswith('TPS') and filename.endswith('.jpg') and ('omr1' in filename or 'omr2' in filename or 'omr3' in filename):
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)
        height, width, _ = image.shape
        aspect_ratio = width / height
        new_height = 500
        new_width = int(new_height * aspect_ratio)
        image = cv2.resize(image, (new_width, new_height))

        # Ubah gambar menjadi abu-abu dan terapkan thresholding
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

        kernel = np.ones((5, 5), np.uint8)
        closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
        contours, _ = cv2.findContours(closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        num_columns = 3
        num_rows = 10
        answers = [[] for _ in range(num_columns)]
        expected_area = 1000

        for contour in contours:
            area = cv2.contourArea(contour)
            if area > expected_area:
                x, y, w, h = cv2.boundingRect(contour)
                column_index = int(round(x / (gray.shape[1] / num_columns)))
                row_index = int(round(y / (gray.shape[0] / num_rows)))

                # Periksa apakah indeks kolom berada dalam jangkauan yang valid
                if 0 <= column_index < len(answers):
                    if len(answers[column_index]) == 0:
                        answers[column_index].append(row_index)
                    else:
                        last_idx = answers[column_index][-1]
                        try:
                            if area > cv2.contourArea(contours[last_idx]):
                                answers[column_index] = [row_index]
                            elif row_index != last_idx:
                                answers[column_index].append(row_index)
                        except IndexError:
                            # Tangani kasus di mana last_idx berada di luar jangkauan kontur
                            print(f"Error processing contour for {filename}: index {last_idx} out of range.")
                else:
                    print(f"Indeks kolom {column_index} berada di luar jangkauan yang valid untuk {filename}")

        # Tambahkan data ke DataFrame
        data['TPS'].append(filename)
        data['Column 1'].append(answers[0])
        data['Column 2'].append(answers[1])
        data['Column 3'].append(answers[2])

# Buat DataFrame
df = pd.DataFrame(data)
df
```

### Mengedit dan Menyimpan Dataframe

```python
# Mengganti nilai yang >2 digit menjadi 0
import pandas as pd

def replace_list_with_zero_if_more_than_one(x):
    # Jika x adalah list dan jumlah elemennya lebih dari satu, gantikan dengan 0
    if isinstance(x, list) and len(x) > 1:
        return 0
    else:
        return x

# Terapkan ke setiap elemen dalam DataFrame
df = df.applymap(replace_list_with_zero_if_more_than_one)
df
```

```python
# Mendefinisikan function untuk mengganti nilai kosong dengan 0
def replace_empty_list_with_zero(x):
    if isinstance(x, list) and len(x) == 0:
        return 0
    else:
        return x

# Terapkan function ke setiap elemen dalam DataFrame
df = df.applymap(replace_empty_list_with_zero)
df
```

```python
# Ekstrak bagian 'omr' dari kolom 'TPS'
df['omr_numeric'] = df['TPS'].str.extract(r'omr(\d+)').astype(int)

# Urutkan DataFrame berdasarkan kolom 'omr_numeric' kemudian oleh kolom 'TPS'
df_sorted = df.sort_values(by=['omr_numeric', 'TPS'])

# Hapus kolom sementara 'omr_numeric'
df_sorted.drop('omr_numeric', axis=1, inplace=True)

# Reset indeks setelah pengurutan
df_sorted.reset_index(drop=True, inplace=True)

df_sorted
```

```python
# Mengganti nilai 10 menjadi 0
def replace_ten_with_zero(cell):
    # Mengganti nilai 10 dengan 0 dalam sel jika sel tersebut merupakan list yang mengandung [10]
    if isinstance(cell, list) and cell == [10]:
        return 0
    return cell

# Terapkan fungsi tersebut ke setiap sel di dataframe
for column in df_sorted.columns:
    df_sorted[column] = df_sorted[column].apply(replace_ten_with_zero)

df_sorted
```

```python
# Sorting dataframe berdasarkan alfabet
df=df_sorted
df
```

```python
# Koreksi fungsi penggantian nama untuk mengecualikan nomor baris dan "-kebalik"
def correct_rename_files(file_name):
    # Hapus angka dan spasi yang mungkin muncul dari nomor baris di awal nama file
    file_name = file_name.lstrip('0123456789 ')
    if '-kebalik' in file_name:
        # Ganti '-kebalik' dengan karakter kosong
        file_name = file_name.replace('-kebalik', '')
    return file_name

# Terapkan function
df['TPS'] = df['TPS'].apply(correct_rename_files)
```

```python
df
```

```python
# Memperbaiki format penamaan pada kolom TPS

# Hilangkan ekstensi .jpg dari nilai-nilai dalam kolom 'TPS'
df['TPS'] = df['TPS'].str.replace('.jpg', '', regex=False)

def concatenate_digits(row):
    # Hapus kurung kotak dari setiap kolom, gabungkan, dan konversi ke integer
    digits = ''.join(filter(str.isdigit, f"{row['Column 1']}{row['Column 2']}{row['Column 3']}"))
    return int(digits) if digits else 0

# Terapkan fungsi concatenate_digits
df['votes'] = df.apply(concatenate_digits, axis=1)

# Ekstrak informasi TPS dan omr
df['TPS_extracted'] = df['TPS'].str.extract(r'(TPS_\d+)')[0]
df['omr'] = df['TPS'].str.extract(r'_(omr\d)')[0]

# Hapus kolom-kolom angka asli
df.drop(['Column 1', 'Column 2', 'Column 3', 'TPS'], axis=1, inplace=True)

# Ganti nama kolom 'TPS_extracted' kembali ke 'TPS' setelah menghapus kolom 'TPS' asli
df.rename(columns={'TPS_extracted': 'TPS'}, inplace=True)

# Pivot dataframe untuk membuat kolom untuk setiap omr dengan TPS sebagai baris
df_pivot = df.pivot(index='TPS', columns='omr', values='votes').fillna(0).astype(int)

# Ganti nama kolom sesuai format yang diinginkan
df_pivot.columns = [f'paslon 0{col[-1]}' for col in df_pivot.columns]

# Reset indeks untuk mendapatkan 'TPS' sebagai kolom bukan sebagai indeks
df_transformed = df_pivot.reset_index()

# Tampilkan DataFrame yang telah diubah
print(df_transformed)
```
